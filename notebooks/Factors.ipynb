```python
def compare_sum_to_num(limit):
    x = list(range(1, limit))
    y = [aliquot_sum(n) for n in x]

    plt.plot(x, y)
    plt.xlabel("number")
    plt.ylabel("aliquot sum")
    plt.plot(x,x)
    plt.show()
```

```python
compare_sum_to_num(100)
```
